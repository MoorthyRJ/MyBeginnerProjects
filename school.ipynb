```python
#creating school using OOPS
import json
from abc import ABC,abstractmethod
class Upload:
  def __init__(self,filename):
    self.filename=filename

  def upload(self,data):
    with open(self.filename,"w") as f:
      json.dump(data,f,indent=4)

  def load(self):
    try:

      with open(self.filename,"r") as f:
       return json.load(f)
    except FileNotFoundError:
      return {}



class Person(ABC):

  def __init__(self,name,age):
    self.name=name
    self.age=age

  @abstractmethod
  def getinfo(self):
    print(f"Person is {self.name} and age is {self.age}")

class Student(Person):
  def __init__(self,name,age,rollno,dept):
    super().__init__(name,age)
    self.rollno=rollno
    self.dept=dept

    self.markslist=Upload(f"{self.rollno}_marks.json")
    self.attendancelist=Upload(f"{self.rollno}_attendancelist.json")

    self.__attendance=self.attendancelist.load()
    self.__marks=self.markslist.load()
    self.__cgpa=0

  def getmark(self):
    if not self.__marks:
      return "No marks uploaded"

    else:
      self.markslist.upload(self.__marks)
      return self.__marks

  def getattendance(self):
    if not self.__attendance:
      return "No attendance uploaded"
    else:

      self.attendancelist.upload(self.__attendance)
      return self.__attendance

  def getcgpa(self):
    return self.__cgpa


  def _uploadmark(self,subject,mark,authorization=None):
    if not isinstance(authorization,Staff):
      print("Unauthorized access")
    else:
        self.__marks[subject]=mark

  def _uploadattendance(self,sub,attendance,authorization=None):
    if not isinstance(authorization,Staff):
      print("Unauthorized access")
    else:
      self.__attendance[sub]=attendance

  def optain_cgpa(self):
    if not self.__marks:
      return "No marks uploaded."

    totalmark=sum(self.__marks.values())

    self.__cgpa=totalmark/len(self.__marks)
    return self.__cgpa

  def getpercentage(self):
    return self.getcgpa()*0.905


  def getinfo(self):
    print(f"RollNo:{self.rollno},Attendence:{self.getattendance()},Marks:{self.getmark()},CGPA:{self.getcgpa():.2f}({self.getpercentage():.2f}).")
    return


class Staff(Person):
  def __init__(self,name,age,empid,dept):
    super().__init__(name,age)
    self.empid=empid
    self.dept=dept


  def getsalary(self):
    return self.__salary

  def getworkexperience(self):
    return self.__workexperience

  def getinfo(self):
    return f'EmpId:{self.empid}'
class Teacher(Staff):
  def __init__(self,name,age,empid,dept,subject):
    super().__init__(name,age,empid,dept)
    self.subject=subject
    self.__salary=10000
    self.__workexperience=5

  def getsalary(self):
    return self.__salary*(self.__workexperience/2)

  def getworkexperience(self):
    return self.__workexperience


  def makeattendence(self,schoolname):
    student=None
    studentname=input("Enter the student name:").strip()
    while True:
      try:
        attendance=int(input("Enter the attendance:"))
        if attendance<0 or attendance>100:
          print("Enter a valid attendance")
          continue
        else:
          break
      except ValueError:
        print("Enter a valid attendance")
    for stud in schoolname.students:
        if stud.name.lower() == studentname.lower():
            student = stud
            break
    if student:
        student._uploadattendance(self.subject, attendance, authorization=self)


        print("Attendance uploaded successfully")
        return
    else:
          print("Student not found")
          return

  def makemark(self,schoolname):
    print(f"Subject:{self.subject}")
    name=input("Enter the student name:").strip()
    while True:
      try:
        mark=int(input("Enter the mark:"))
        if mark<0 or mark>100:
          print("Enter a valid mark")
          continue
        else:
          break
      except ValueError:
        print("Enter a valid mark")




    student=None
    for s in schoolname.students:
      if s.name.lower()==name.lower():
        student=s
        break

    if student:
      student._uploadmark(self.subject,mark,authorization=self)
      print("Mark uploaded successfully")
      return
    else:
        print("Student not found")

        return


  def getinfo(self):
    return f'Subject:{self.subject}'

class School():
  def __init__(self,name):
    self.name=name
    self.staffs=[]
    self.students=[]

  def addstaff(self,staff):
    self.staffs.append(staff)

  def addstudent(self,student):
    self.students.append(student)
  def allstaff(self):
    for staff_member in self.staffs:
      if isinstance(staff_member,Teacher):

         print(f"{staff_member.name}-{staff_member.subject}")
      else:

        print(f"{staff_member.name}--{staff_member.dept}")

  def allstudent(self):
    for student in self.students:
      print(f"{student.name}-{student.rollno}-{student.dept}")
    return
amc=School("The American College")
moorthy=Student("Moorthy",21,"23CIT40","CIT")
virgil=Student("Virgil",20,"23CIT86","CIT")
amc.addstudent(moorthy)
amc.addstudent(virgil)
raman=Teacher("Raman",35,"ACC1","CIT","Accountancy")
taj=Teacher("Taj",30,"ACC2","CIT","Cost acc")

amc.addstaff(raman)
amc.addstaff(taj)

amc.allstudent()
taj.makeattendence(amc)
raman.makeattendence(amc)

moorthy.optain_cgpa()
print(virgil.getattendance())


moorthy.getinfo()
```
