```python
import random




print("You only have 5 changes")
option=random.randint(1,100+1)
while True:
  try:

    guess=int(input("Enter Your Guess(1-100):"))
    if guess<=0 or guess>100:
      print("Guess can't be negative or more than 100")
    else:
      break
  except ValueError:
    print("Guess  must be a number between 1-100")



chance=4



while True:
      print(f"{chance} chances left.")
      if guess==option:
        print(f"You got it!! with {chance} left. ")
        break
      if chance>=0:
        try:
          if guess>option:
            guess=int(input("Too high:"))
            while True:
              if guess<=0 or guess>100:
                print("Guess can't be negative or more than 100")
              else:
                break


          elif guess<option:
            while True:
              guess=int(input("Too low:"))
              if guess<=0 or guess>100:
                print("Guess can't be negative or more than 100")
              else:
                break


          chance-=1


        except ValueError:
          print("Guess  must be a number between 1-100")
      else:
        print(f"You Lost.The correct answer is{option}")
        break






```
