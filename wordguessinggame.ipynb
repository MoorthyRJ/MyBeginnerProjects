```python
#wordguessing game
import random
import listforwordgame as words





def displayanswer(answer):
  print(" ".join(answer))



def displayhint(hint):
  print(" ".join(hint))

def main():
  print(" "*50)
  print("<welcome to wordguess game>")
  print(" "*50)

  answer=random.choice(words.words)
  guessedword=set()
  hint=['_']*len(answer)
  noofguess=10


  while noofguess>0:
    print(f"you only have {noofguess} left.")
    displayhint(hint)
    guess=input("Enter your Guess word:").lower()

    if len(guess)!=1 or not guess.isalpha():
      print("please enter a single alphabet")
      continue
    if guess in guessedword:
      print("you have already guessed this word")

      continue
    guessedword.add(guess)
    if guess in answer:
      for i in range(len(answer)):
        if answer[i]==guess:
          hint[i]=guess

      noofguess-=1


    else:
      print("No in the word!!")
      noofguess-=1

    if "_" not in hint:
        print("you won")
        displayanswer(answer)
        break


  if noofguess==0:
    print("you lost")
    displayanswer(answer)



if __name__=="__main__":
  main()







```
