```python
#todolist
import json

todo=[]
def displaymsg():

  print("Press 1 to view Todo lsit:")
  print("Press 2 to add Todo:")


def viewtodo(todo):
  if not todo:
    print("No todo yet.Lets Create one")
    t=input("Enter the Todo:")
    todo.append(t)
    return


  for i,t in enumerate(todo,1):
    print(f"{i}.{t}.")
  a=input("want to mark done any todo(yes/no):").lower()
  if a =="yes":
    marktodo(todo)

  else:
    return



def addtodo(todo):

    t=input("Enter the Todo:")
    todo.append(t)
    print(f"{t} Added Sucessfully")


    return

def savetodo(todo,filename="todolistsave.json"):
  with open(filename,"w")as f:
    json.dump(todo,f,indent=4)
  return

def loadtodo(filename="todolistsave.json"):
  try:
    if filename:
      with open(filename,"r")as f:
        todo=json.load(f)
      return todo
    else:
      return[]
  except FileNotFoundError:
    return[]
def marktodo(todo):

  while True:
    mark=int(input("Enter the number of todo mark as done."))
    if 0< mark<len(todo):
      mark=mark-1
      todo[mark]=todo[mark]+" ✓"
      print("Marked as done")

    else:
      print("Enter a correct value:")

    break


    return

def main():
  todo=loadtodo()
  print("<Welcome to ToDo list>")
  print(' '*50)


  while True:
      displaymsg()

      choice=input("What you want to do (q to quit):").lower()
      if choice=='1':
        viewtodo(todo)
        savetodo(todo)

      elif choice=="2":
        addtodo(todo)
        savetodo(todo)
      elif choice =="q":
        break
      else:
        print("Invalid input")




main()



```
